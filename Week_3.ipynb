The statement `x = np.linspace(0, 10, 100)` creates a **NumPy array** with 100 evenly spaced values between 0 and 10, inclusive.

---

### Explanation of the Code

```python
import numpy as np

# Generate 100 evenly spaced numbers between 0 and 10
x = np.linspace(0, 10, 100)

# Display the array
x
```

---

### Breakdown:
1. **`np.linspace(start, stop, num)`**:
   - **`start`**: The starting value of the sequence. Here, it is `0`.
   - **`stop`**: The ending value of the sequence. Here, it is `10`.
   - **`num`**: The number of values to generate. Here, it is `100`.

2. **Key Points**:
   - Unlike `np.arange`, which generates values with a fixed step, `np.linspace` generates a specified number of evenly spaced points between `start` and `stop`.
   - The resulting array includes both the `start` and `stop` values (inclusive).

3. **Output**:
   The array will look like this:
   ```
   array([ 0.        ,  0.1010101 ,  0.2020202 , ...,  9.7979798 ,  9.8989899 , 10.        ])
   ```
   It contains 100 evenly spaced values, including `0` and `10`.

---

### Run the cell in your Jupyter Notebook:

```python
x = np.linspace(0, 10, 100)
x
```

Output:
```
array([ 0.        ,  0.1010101 ,  0.2020202 , ...,  9.7979798 ,  9.8989899 , 10.        ])
```

---

### Additional Examples:

1. **Change the Number of Points**:
   ```python
   x = np.linspace(0, 10, 5)
   print(x)
   ```
   Output:
   ```
   [ 0.   2.5  5.   7.5 10. ]
   ```

2. **Include or Exclude the Endpoint**:
   By default, `np.linspace` includes the endpoint (`10` in this case). To exclude it:
   ```python
   x = np.linspace(0, 10, 5, endpoint=False)
   print(x)
   ```
   Output:
   ```
   [0. 2. 4. 6. 8.]
   ```

3. **Plot the Values**:
   ```python
   import matplotlib.pyplot as plt

   # Generate values
   x = np.linspace(0, 10, 100)
   y = np.sin(x)  # Sine wave

   # Plot
   plt.plot(x, y)
   plt.title("Sine Wave")
   plt.xlabel("x")
   plt.ylabel("sin(x)")
   plt.show()
   ```

This will create a smooth sine wave because `np.linspace` ensures evenly spaced values.

 ## **6. Customizing Plots**
 ### Example: Adding Multiple Plots in a Single Figure

### Explanation of the Code

1. **`x = np.linspace(0, 10, 100)`**:
   - This generates 100 evenly spaced values between 0 and 10, inclusive.
   - The resulting array `x` represents the input values (independent variable) for the sine and cosine functions.
   - These values are in radians since `np.sin` and `np.cos` use radian input.

---

2. **`y1 = np.sin(x)`**:
   - **Purpose**: Computes the sine of each value in `x`.
   - The sine function maps each value in `x` to a new value in the range `[-1, 1]`.
   - Result: `y1` is a NumPy array of the same size as `x`, containing the sine values for each corresponding element in `x`.

   - Example behavior:
     - When `x = 0`, `sin(0) = 0`.
     - When `x = /2` (approximately 1.57), `sin(/2) = 1`.
     - When `x = ` (approximately 3.14), `sin() = 0`.

---

3. **`y2 = np.cos(x)`**:
   - **Purpose**: Computes the cosine of each value in `x`.
   - The cosine function maps each value in `x` to a new value in the range `[-1, 1]`.
   - Result: `y2` is a NumPy array of the same size as `x`, containing the cosine values for each corresponding element in `x`.

   - Example behavior:
     - When `x = 0`, `cos(0) = 1`.
     - When `x = /2` (approximately 1.57), `cos(/2) = 0`.
     - When `x = ` (approximately 3.14), `cos() = -1`.

---

### Relationships Between `y1` and `y2`:
- **Sine and Cosine Waves**:
  - Both `y1` and `y2` represent periodic wave functions, with a period of `2` (approximately 6.28).
  - The sine wave (`y1`) starts at 0, increases to 1, decreases back to 0, then goes to -1, and repeats.
  - The cosine wave (`y2`) starts at 1, decreases to 0, goes to -1, and repeats.

- **Phase Difference**:
  - Sine and cosine waves are identical except for a phase shift:
    - The sine wave lags the cosine wave by `/2` radians (90 degrees).

---

### Resulting Data:
After executing these lines:
- `x`: Contains 100 evenly spaced values between 0 and 10.
- `y1`: Contains the sine of each value in `x`.
- `y2`: Contains the cosine of each value in `x`.

These arrays can be used for mathematical analysis, graphing, or any application involving trigonometric functions.

````python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plotting both sine and cosine on the same figure
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="Sine Wave", color="blue")
plt.plot(x, y2, label="Cosine Wave", color="orange")
plt.title("Sine and Cosine Waves", fontsize=16)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Y Values", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
````

 # 3: Using Seaborn for Advanced Visualizations
 ## Goal:
 - Leverage Seaborn's capabilities for creating visually appealing and advanced plots.

 ---
 ## **1. Setting Up Seaborn**

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Using a built-in Seaborn style
sns.set_theme(style="whitegrid")

# Generating a dummy dataset for demonstration
np.random.seed(42)
data = pd.DataFrame({
    "Feature_A": np.random.normal(50, 10, 100),
    "Feature_B": np.random.normal(30, 5, 100),
    "Category": np.random.choice(["Group 1", "Group 2", "Group 3"], 100),
    "Outcome": np.random.choice([0, 1], 100)
})

````

 ---
 ## **2. Scatter Plots**
 ### Enhanced Scatter Plot with `sns.scatterplot`

````python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=data,
    x="Feature_A",
    y="Feature_B",
    hue="Category",
    style="Outcome",
    palette="deep",
    s=100
)
plt.title("Enhanced Scatter Plot: Feature A vs. Feature B", fontsize=16)
plt.xlabel("Feature A", fontsize=12)
plt.ylabel("Feature B", fontsize=12)
plt.legend(title="Category & Outcome")
plt.show()
````

 ---
 ## **3. Box Plots**
 ### Box Plot Showing Distributions and Outliers

````python

plt.figure(figsize=(10, 6))
sns.boxplot(
    data=data,
    x="Category",
    y="Feature_A",
    palette="pastel"
)
plt.title("Box Plot: Feature A Distribution by Category", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Feature A", fontsize=12)
plt.show()
````


 ---
 ## **4. Heatmaps**
 ### Visualizing Correlations with a Heatmap

 Creating a correlation matrix

````python
correlation_matrix = data[["Feature_A", "Feature_B", "Outcome"]].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    square=True,
    cbar_kws={"shrink": 0.8}
)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
````


 ---
 ## **5. Pair Plots**
 ### Exploring Relationships Between Features with `sns.pairplot`


````python
sns.pairplot(
    data=data,
    vars=["Feature_A", "Feature_B"],
    hue="Category",
    palette="dark",
    diag_kind="kde"
)
plt.suptitle("Pair Plot: Feature Relationships by Category", y=1.02, fontsize=16)
plt.show()
````

````python
# ---
# title: "Final Project: Visualizing and Analyzing a Dataset"
# ---

# ## **Section 5: Final Project Exercise**

# ### Goal:
# Apply visualization techniques to explore and present insights from a dataset.

# ---
# ## **Step 1: Load and Explore the Dataset**
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load a sample dataset: Iris dataset
from sklearn.datasets import load_iris
iris_data = load_iris(as_frame=True)
df = iris_data.frame

# Display basic info
print("Dataset Preview:")
print(df.head())

# Display summary statistics
print("\nSummary Statistics:")
print(df.describe())

# ---
# ## **Step 2: Create Visualizations**

# ### 1. Histogram: Explore the distribution of sepal length
plt.figure(figsize=(8, 6))
sns.histplot(df['sepal length (cm)'], kde=True, color="blue")
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# ### 2. Heatmap: Visualize correlations
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# ### 3. Scatter Plot: Relationship between petal length and petal width, with hue
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="petal length (cm)",
    y="petal width (cm)",
    hue=iris_data.target,
    palette="viridis"
)
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title="Species", labels=iris_data.target_names)
plt.show()

# ---
# ## **Step 3: Combine Visualizations into Subplots**
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Subplot 1: Histogram
sns.histplot(df['sepal length (cm)'], kde=True, color="blue", ax=axes[0, 0])
axes[0, 0].set_title("Distribution of Sepal Length")
axes[0, 0].set_xlabel("Sepal Length (cm)")
axes[0, 0].set_ylabel("Frequency")

# Subplot 2: Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5, ax=axes[0, 1])
axes[0, 1].set_title("Correlation Heatmap")

# Subplot 3: Scatter Plot
sns.scatterplot(
    data=df,
    x="petal length (cm)",
    y="petal width (cm)",
    hue=iris_data.target,
    palette="viridis",
    ax=axes[1, 0]
)
axes[1, 0].set_title("Petal Length vs Petal Width")
axes[1, 0].set_xlabel("Petal Length (cm)")
axes[1, 0].set_ylabel("Petal Width (cm)")

# Empty subplot for symmetry
axes[1, 1].axis("off")

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig("combined_visualizations.png", dpi=300, bbox_inches="tight")
plt.show()

# ---
# ## **Conclusion**
# - Loaded and explored the Iris dataset.
# - Created three visualizations: histogram, heatmap, and scatter plot.
# - Combined visualizations into a subplot layout for comprehensive analysis.
# - Saved the combined figure as an image file.


````
